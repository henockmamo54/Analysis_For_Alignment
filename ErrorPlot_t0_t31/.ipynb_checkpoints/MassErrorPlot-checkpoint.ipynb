```python
import numpy as np
import pandas as pd
import os
import seaborn as sns
from matplotlib import pyplot as plt

import warnings

warnings.simplefilter("ignore")
```

```python
path="data/" 
filelist=os.listdir(path) 
data=pd.DataFrame()
```

```python
for  p in filelist[0:]: 
    temp=pd.read_csv(path+p)
    data=pd.concat([data,temp])

# filelist
```

```python
plt.rc('font', size=14) 
plt.figure(figsize=(12,12))
x=np.arange(data.t0_error.min(),data.t0_error.max())

plt.scatter(data.t0_error,data.t31_error,c='k',marker='o',alpha=0.5,s=200)
plt.vlines(x=0,ymin=-data.t31_error.max(),ymax=data.t31_error.min(),color='red',linestyle="-.",linewidth=4)
plt.hlines(y=0,xmin=data.t0_error.max(),xmax=data.t0_error.min(),color='m',linestyle="-.",linewidth=4)

plt.plot(x,2*x,linestyle="-.",linewidth=4)
plt.plot(x,0.5*x,linestyle="-.",linewidth=4) 
plt.plot(x,x,linestyle="-.",linewidth=4) 
plt.ylim([-2200,100])

plt.xlabel("Mass error (ppm), Unlabeled sample")
plt.ylabel("Mass error (ppm), Labeled sample")

plt.savefig(f"Error Plot.jpeg",dpi=900, bbox_inches = "tight")
```

```python
tempdata=data
tempdata["diff"]=abs((tempdata.t0_error - tempdata.t31_error)/tempdata.t31_error)
tempdata[(tempdata.t0_error<-200) & ((tempdata.t31_error<-200))].sort_values(by="diff",ascending=True).head(15)
```

```python
print(tempdata[tempdata.t31_error>-200].shape[0]/tempdata.shape[0]) #t31=0
print(tempdata[tempdata.t0_error>-200].shape[0]/tempdata.shape[0]) # t0=0
print(tempdata[(tempdata.t0_error<-200) & ((tempdata.t31_error<-200)) & (tempdata["diff"]>0.95)].shape[0]/tempdata.shape[0]) # t0 = 2*t31
print(tempdata[(tempdata.t0_error<-200) & ((tempdata.t31_error<-200)) & (tempdata["diff"]<0.15)].shape[0]/tempdata.shape[0]) # t0 = t31
print(tempdata[(tempdata.t0_error<-200) & ((tempdata.t31_error<-200)) & 
               (tempdata["diff"]<0.65) & (tempdata["diff"]>0.45)].shape[0]/tempdata.shape[0]) # t0 = 0.5*t31
```
