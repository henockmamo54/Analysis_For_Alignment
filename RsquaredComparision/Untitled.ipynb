```python
import os
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
noAlinoZero=[6120,9024,11612,12321]
noAlin=[6463,9895,13036,14038]
Alin=[6795,10277,13537,14524]
```

```python
plt.figure(figsize=(9,6))
bar_width = 0.25
index=np.array([0,1,2,3])

# '#3CAEA3','#F6D55C','#ED553B','#20639b'

plt.bar(index,noAlinoZero,bar_width, label= "With out Zero Ionscore" ,color='#ED553B')
plt.bar(index+bar_width,noAlin,bar_width, label= "With Zero Ionscore",color='#F6D55C')
plt.bar(index+2*bar_width,Alin,bar_width, label="Alignment",color='#20639b')

plt.xticks([0,1,2,3],[r'$R^2$ >= 95',r'$R^2$ >= 90',r'$R^2$ >= 80',r'$R^2$ >= 75'])
plt.legend()
```
