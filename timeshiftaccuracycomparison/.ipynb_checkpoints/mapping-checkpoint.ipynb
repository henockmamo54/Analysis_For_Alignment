```python

path= 'data_0.5'
filelist=os.listdir(path)       

data=pd.DataFrame()
for  p in filelist:
    d=pd.read_csv(os.path.join(path,p))
    d["protein"]=p.replace(".csv","").replace("mid_ret_time_","")
    data=pd.concat([data,d])       

print(data.shape)
data=data[data.Abun> 0.5E8]
print(data.shape)
data.head()
```

```python
points=[]
centers=[]
times=[0, 1, 2, 3, 6, 7, 9, 13, 16, 21, 24, 31]
for r in range(data.shape[0]):
    temp=[]
    closest_align_t0=pd.NA
    
    for t in times:
        if(t==0):
            closest_align_t0=float(closest_value(Finaldata.L0, data.loc[r,f"t{t}"]))
            
        if(data.loc[r,f"i{t}"]>0 and data.loc[r,f"t{t}"] <121 and data.loc[r,f"t{t}"] >=0 ):
            points.append([data.loc[r,f"t{t}"],data.loc[r,f"m{t}"],t,
                           float(np.array(Finaldata[[f"L{t}"]])[list(Finaldata.L0).index(closest_align_t0)]),
                          closest_align_t0,r])
            temp.append(points[-1])
            
    temp=np.array(temp)
    centers.append([ np.median(temp[:,0]), np.median(temp[:,1]),r])

len(points)
```

```python
dfpoints=pd.DataFrame(points)
dfpoints.columns=["Original","mz","t","after","t0","cluster"]
dfpoints=dfpoints[["mz","t","Original","after","t0","cluster"]]
# dfpoints=dfpoints[(dfpoints.Original<120) & (dfpoints.Original>1) & (dfpoints.after)>1 & (dfpoints.after<120)]
# dfpoints=dfpoints[(dfpoints.Original<120) & (dfpoints.Original>1) & (dfpoints.after<119) & (dfpoints.after>1)   ]
dfpoints=dfpoints[(dfpoints.t0 != 0.01111964) & (dfpoints.t0 != 0.01112) & (dfpoints.t0 !=120.0010)]
dfpoints=dfpoints.reset_index(drop=True)
print(dfpoints.shape)
dfpoints.head()
```

```python
def weightedL2(a,b,w):
    q = a-b
#     print(a,b,q,np.sqrt((w*q*q).sum()))
#     print("\t",np.linalg.norm(a-b))
    return np.sqrt((w*q*q).sum())
```

```python
d_mz=0.7
d_t=0.5

index=2
print(dfpoints.loc[index,"cluster"])

o_t=dfpoints.loc[index,"Original"] #orginal_cluster_center_t
o_mz=dfpoints.loc[index,"mz"] #orginal_cluster_center_mz

tempcenters=dfcenters[(dfcenters.c_t >= o_t-d_t ) & (dfcenters.c_t <= o_t+d_t ) &
                      (dfcenters.c_mz >= o_mz-d_mz ) & (dfcenters.c_mz <= o_mz+d_mz )
                     ].reset_index(drop=True)
dist=[]
for c in range(tempcenters.shape[0]):
    dist.append((weightedL2(np.array([o_t,o_mz]),np.array([tempcenters.loc[c,"c_t"],tempcenters.loc[c,"c_mz"]]),[0.7,0.3])))
newcluster=tempcenters.loc[np.argmin(np.array(dist)),"cluster"]
print(newcluster)
```

```python
d_mz=0.85
d_t=0.5

ytrue=[]
ypred=[]
for index in range(dfpoints.shape[0]): 
#     print(index)
    true_cluster=(dfpoints.loc[index,"cluster"])
    ytrue.append(true_cluster)

    o_t=dfpoints.loc[index,"Original"] #orginal_cluster_center_t
    o_mz=dfpoints.loc[index,"mz"] #orginal_cluster_center_mz

    tempcenters=dfcenters[(dfcenters.c_t >= o_t-d_t ) & (dfcenters.c_t <= o_t+d_t ) &
                      (dfcenters.c_mz >= o_mz-d_mz ) & (dfcenters.c_mz <= o_mz+d_mz )
                     ].reset_index(drop=True)
    
    if(tempcenters.shape[0]==0): 
        count=2
        while(tempcenters.shape[0]==0):
            tempcenters=dfcenters[(dfcenters.c_t >= o_t- count*d_t ) & (dfcenters.c_t <= o_t+ count*d_t ) &
                      (dfcenters.c_mz >= o_mz-count*d_mz ) & (dfcenters.c_mz <= o_mz+ count*d_mz )
                     ].reset_index(drop=True)
#             print("\t",count,tempcenters.shape)
            count+=1
            
    dist=[]
    for c in range(tempcenters.shape[0]):
        dist.append((weightedL2(np.array([o_t,o_mz]),
                                np.array([tempcenters.loc[c,"c_t"],tempcenters.loc[c,"c_mz"]]),
                                [0.7,0.3])))
    newcluster=tempcenters.loc[np.argmin(np.array(dist)),"cluster"]
#     print(true_cluster,newcluster)
    ypred.append(newcluster)
    
```

```python
temp=data[(data.i0>0)&(data.i1>0) & (data.t0<121) & (data.t1<121)][["protein","PeptideSeq","Charge","t0","t1"]]
print(temp.shape)
temp.head()
```

```python
# def closest(lst, K):      
#     return lst[min(range(len(lst)), key = lambda i: abs(lst[i]-K))]
# closest_align_t0=[closest(Finaldata.L0, x) for x in temp.t0 ]

def closest_value(input_list, input_value): 
    arr = np.asarray(input_list) 
    i = (np.abs(arr - input_value)).argmin() 
    return arr[i]

closest_align_t0=[closest_value(Finaldata.L0, x) for x in temp.t0 ]
newtime_align=[ Finaldata.L1[list(Finaldata.L0).index(x)] for x in closest_align_t0]
temp["al"]=newtime_align
print(temp.shape)
temp.head()
```

```python
plt.rc("font",size=14)
plt.figure(figsize=(6,6))
plt.scatter(temp.t0-temp.t1,temp.t0-temp.al,s=100,c='k',alpha=0.5)
plt.xlim([-6,6])
plt.ylim([-0.4,0.4])
# temp.t0-temp.al.astype('float')
plt.xlabel("\u0394t (Actual)")
plt.ylabel("\u0394t (Alignment)")
# plt.savefig(f"delta_t_0_1.jpeg",dpi=900, bbox_inches = "tight")
```
