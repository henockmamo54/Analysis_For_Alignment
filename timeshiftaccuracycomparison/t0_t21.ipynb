```python

path= 'data'
filelist=os.listdir(path)       

data=pd.DataFrame()
for  p in filelist:
    d=pd.read_csv(os.path.join(path,p))
    d["protein"]=p.replace(".csv","").replace("mid_ret_time_","")
    data=pd.concat([data,d])       

print(data.shape)
data=data[data.Abun> 0.5E8]
print(data.shape)
data.head()
```

```python
temp=data[(data.i0>0)&(data.i21>0) & (data.t0<121) & (data.t21<121)][["protein","PeptideSeq","Charge","t0","t21"]]
print(temp.shape)
temp.head()
```

```python
# def closest(lst, K):      
#     return lst[min(range(len(lst)), key = lambda i: abs(lst[i]-K))]
# closest_align_t0=[closest(Finaldata.L0, x) for x in temp.t0 ]

def closest_value(input_list, input_value): 
    arr = np.asarray(input_list) 
    i = (np.abs(arr - input_value)).argmin() 
    return arr[i]

closest_align_t0=[closest_value(Finaldata.L0, x) for x in temp.t0 ]
newtime_align=[ Finaldata.L21[list(Finaldata.L0).index(x)] for x in closest_align_t0]
temp["al"]=newtime_align
print(temp.shape)
temp.head()
```

```python
plt.rc("font",size=14)
plt.figure(figsize=(6,6))
plt.scatter(temp.t0-temp.t21,temp.t0-temp.al,s=100,c='k',alpha=0.5)
plt.xlim([-6,6])
plt.ylim([-1,0.25])
# temp.t0-temp.al.astype('float')
plt.xlabel("\u0394t (Actual)")
plt.ylabel("\u0394t (Alignment)")
plt.savefig(f"delta_t_0_21.jpeg",dpi=900, bbox_inches = "tight")
```
