2 Prepare experimental data
===

```python

path= f'data_0.5/{organ}/'
filelist=os.listdir(path)       

data=pd.DataFrame()
for  p in filelist:
    d=pd.read_csv(os.path.join(path,p))
    d["protein"]=p.replace(".csv","").replace("mid_ret_time_","")
    data=pd.concat([data,d])       

print(data.shape)
data=data[data.Abun> 0.5E9]
print(data.shape) 
data=data[data.NDP>=4] 
data=data.reset_index(drop=True)
print(data.shape)
data.head() 
```

```python
def closest_value(input_list, input_value): 
    arr = np.asarray(input_list) 
    i = (np.abs(arr - input_value)).argmin() 
    return arr[i]
```

3 prepare the ground truth map
===

```python
points=[]
centers=[]
times=[0, 1, 2, 3, 6, 7, 9, 13, 16, 21, 24, 31]
groups=[]

for r in range(data.shape[0]):
    temp=[]
    groupset=set()
    closest_align_t0=pd.NA
    
    for t in times:
        if(t==0):
            closest_align_t0=float(closest_value(Finaldata.L0, data.loc[r,f"t{t}"]))
            
        if(data.loc[r,f"i{t}"]>0 and data.loc[r,f"t{t}"] <121 and data.loc[r,f"t{t}"] >=0 ):
            points.append([data.loc[r,f"t{t}"],data.loc[r,f"m{t}"],t,
                           float(np.array(Finaldata[[f"L{t}"]])[list(Finaldata.L0).index(closest_align_t0)]),
                          closest_align_t0,r])
            temp.append(points[-1])    
            groupset.add(str(data.loc[r,f"t{t}"])+"_"+str(data.loc[r,f"m{t}"])+"_"+str(t))
            
    temp=np.array(temp)
    centers.append([ np.median(temp[:,0]), np.median(temp[:,1]),r])
    groups.append(groupset)

len(points)
```

```python
dfpoints=pd.DataFrame(points)
dfpoints.columns=["Original","mz","t","after","t0","cluster"]
dfpoints=dfpoints[["mz","t","Original","after","t0","cluster"]]
# dfpoints=dfpoints[(dfpoints.Original<120) & (dfpoints.Original>1) & (dfpoints.after)>1 & (dfpoints.after<120)]
# dfpoints=dfpoints[(dfpoints.Original<120) & (dfpoints.Original>1) & (dfpoints.after<119) & (dfpoints.after>1)   ]
dfpoints=dfpoints[(dfpoints.t0 != 0.01111964) & (dfpoints.t0 != 0.01112) & (dfpoints.t0 !=120.0010)]
dfpoints=dfpoints.reset_index(drop=True)
print(dfpoints.shape)

dfpoints.columns=["mz","t","Original","after","t0","cluster"]
dfpoints["id"]=dfpoints.Original.astype("str")+"_"+dfpoints.mz.astype("str")+"_"+dfpoints.t.astype("str")
dfpoints=dfpoints.sort_values(by=["mz","Original"])
dfpoints=dfpoints.reset_index(drop=True)
dfpoints.head() 
```

4 Prepare new groups(tools) based on m/z and delta t
===

```python
d_mz=0.5
d_t=0.25
visited=[]
bg=[]

for r in range(dfpoints.shape[0]): #dfpoints.shape[0]
    print(r) 
    
    tempt,tempmz,temptp,tempid=dfpoints.loc[r,["Original","mz","t","id"]]
    groupset=set()
    
    if(tempid not in visited):   
        
        tempclosepoints=dfpoints[(dfpoints.Original <= tempt+d_t ) & (dfpoints.Original >= tempt-d_t)& 
                                 (dfpoints.mz <= tempmz+d_mz ) & (dfpoints.mz >= tempmz-d_mz ) ].reset_index(drop=True)
        
        
        for k in range( tempclosepoints.shape[0]):  
            
            tempid1=tempclosepoints.id[k]
            if(tempid1 not in visited):
                groupset.add(tempid1)
                visited.append(tempid1)
        bg.append(groupset)
```

```python
d_mz=0.5
d_t=0.25
visited=[]
af=[]

for r in range(dfpoints.shape[0]): #dfpoints.shape[0]
    print(r) 
    
    tempt,tempmz,temptp,tempid=dfpoints.loc[r,["Original","mz","t","id"]]
    groupset=set()
    
    if(tempid not in visited):   
        
        tempclosepoints=dfpoints[(dfpoints.after <= tempt+d_t ) & (dfpoints.after >= tempt-d_t)& 
                                 (dfpoints.mz <= tempmz+d_mz ) & (dfpoints.mz >= tempmz-d_mz ) ].reset_index(drop=True)
        
        
        for k in range( tempclosepoints.shape[0]):  
            
            tempid1=tempclosepoints.id[k]
            if(tempid1 not in visited):
                groupset.add(tempid1)
                visited.append(tempid1)
        af.append(groupset)
```

```python
gt=[x for x in groups if len(x)>1]
woa=[x for x in bg if len(x)>1]
wa=[x for x in af if len(x)>1]
```

```python
count=0
sum_n_u=0
sum_n_mgi=0
for i in range(len(gt)): #len(gt)
    vals=(([x for x in wa if(len(x.intersection(gt[i]))>0)]))
    if(len(vals)>0):
#         print(i,len(vals))
        
        n=sum([len(x.intersection(gt[i])) for x in vals])
        u=sum([len(x) for x in vals])
        mgi=len(gt[i])*len(vals)
        count+=1
        sum_n_u +=n/u
        sum_n_mgi+=n/mgi
        
        precision=(1/count)*sum_n_u
        recall = (1/count)*sum_n_mgi
        
print("precision: ",precision," recall: ",recall)
print("F1-score: ", (2*precision*recall)/(precision+recall))    
```
