```python
TN=0
FN=0
FP=0
TP=0
start=12 
elu_window=0

temp_fillalldata=pd.read_csv(filepath+files[files.index('6PGD1_YEAST.Quant.csv') ],skiprows=3,index_col=False)  
temp_fillalldata.fillna(-500)

for row in range(temp_fillalldata.shape[0]):#temp_fillalldata.shape[0]
    
    expriments=[]
    for i in range(40):
        try:
            expriments.append(list(temp_fillalldata.iloc[row,start +i*16:(start+16)+ i*16].astype('float')))
        except:
            expriments.append([-500]*16)
            
        if(i>19 and (expriments[-1][1])>0 and (sum(expriments[-1][5:11])>0) ): # check ionscore, to compute elution window 
            elu_window=abs(expriments[-1][11]-expriments[-1][12])

    for i in range(20):
        temp=expriments[i]
        isotopsum=sum(temp[5:11])

        if( isotopsum ==0 or isotopsum < 1E5 or abs(temp[11]-temp[12])<= 0.5*elu_window  ): #
            TN+=1
        elif(isotopsum>0):
            FP+=1

print(TN,FP,TP,FN) 
```

```python
TN=0
FN=0
FP=0
TP=0
start=12 

for f in range(len(files)):#
    try:
        print(f,files[f])
        
        elu_window=0
        
        #read quant file
        try:
            temp_fillalldata=pd.read_csv(filepath+files[f],skiprows=3,index_col=False)  
            temp_fillalldata.fillna(-500)
        except:
            temp_fillalldata=pd.read_csv(filepath+files[f],index_col=False,skiprows=1)
            temp_fillalldata.fillna(-500)

            
        for row in range(temp_fillalldata.shape[0]):#temp_fillalldata.shape[0]

            expriments=[] #acculate the peptied experiments result as a list
            for i in range(40):
                try:
                    expriments.append(list(temp_fillalldata.iloc[row,start +i*16:(start+16)+ i*16].astype('float')))
                except:
                    expriments.append([-500]*16)

                if(i>19 and (expriments[-1][1])>0 and (sum(expriments[-1][5:11])>0) ): # check ionscore, to compute elution window 
                    elu_window=abs(expriments[-1][11]-expriments[-1][12])

            # count the first 20 experiments
            for i in range(20):
                temp=expriments[i]
                isotopsum=sum(temp[5:11])

                if( isotopsum ==0  or abs(temp[11]-temp[12])<= 0.5*elu_window  ): #or isotopsum < 1E5
                    TN+=1
                elif(isotopsum>0):
                    FP+=1

        print(TN,FP,TP,FN) 
        print("\t\t--->\t", (100*FP/(TN+FP)))
    except:
        print("======================>E",files[f])
```
