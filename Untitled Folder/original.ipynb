```python
TN=0
FN=0
FP=0
TP=0
start=12

nonzeropoints=0
elu_window=0

temp_fillalldata=pd.read_csv( filepath+files[0],skiprows=3,index_col=False)  
temp_fillalldata.fillna(-500)

for row in range(temp_fillalldata.shape[0]):
    #read experiments
    expriments=[]
    for i in range(40):
        try:
            expriments.append(list(temp_fillalldata.iloc[row,start +i*16:(start+16)+ i*16].astype('float')))
        except:
            expriments.append([-500]*16)
        if(i>19 and (expriments[-1][1])>0): # check ionscore
            nonzeropoints+=1 # count nonzero ionscore points
            if(sum(expriments[-1][5:11])>0): # to compute elution window
                elu_window=abs(expriments[-1][11]-expriments[-1][12])

    if nonzeropoints >=4:
        for i in range(20):
            temp=expriments[i]
            if(temp[1]==100 or temp[1]==0):#ionscore ==0
                isotopsum=sum(temp[5:11])
                if( isotopsum ==0 or isotopsum < 1E5 or abs(temp[11]-temp[12])<= 0.5*elu_window):
                    TN+=1
                elif(isotopsum>0):
                    FP+=1

        for i in range(20,40):
            temp=expriments[i]
            if(temp[1]==0): #ionscore ==0
                isotopsum=sum(temp[5:11])
                if( isotopsum > 0 and isotopsum > 1E5 and abs(temp[11]-temp[12])>= 0.5*elu_window):
                    TP+=1
                elif(isotopsum==0 or isotopsum < 1E5 or abs(temp[11]-temp[12])<= 0.5*elu_window):
                    FN+=1
print(TN,FP,TP,FN)
```
