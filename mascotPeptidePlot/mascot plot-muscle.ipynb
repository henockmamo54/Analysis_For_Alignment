```python
import numpy as np
import pandas as pd
```

```python
data1=[5149,4650,4774,4519,4039,4306,4522,3768,3254,3491,2877,2445]
# data2=[10840,10034,10213,9916,9043,9431,9926,8944,7845,8409,7500,6714]
data2=[9839,9057,9395,9031,8141,8318,8647,8184,6599,7483,6607,5854]
```

```python
from matplotlib import pyplot as plt
plt.figure(figsize=(9,6))
bar_width = 0.7
index=np.array([0,1,2,3,6,7,9,13,16,21,24,31])

# '#3CAEA3','#F6D55C','#ED553B','#20639b'

plt.bar(index,data1,bar_width, label= "With out Alignment" ,color='#F6D55C') #ED553B 
plt.bar(index+bar_width+0.1,data1,bar_width, label="With Alignment",color='#20639b')

plt.xticks([0,1,2,3,6,7,9,13,16,21,24,31],[0,1,2,3,6,7,9,13,16,21,24,31])
plt.legend()

# plt.savefig(f"{organ}.jpeg",dpi=900, bbox_inches = "tight")
```

```python
from matplotlib import pyplot as plt
plt.figure(figsize=(40,15))
plt.rc('font', size=14) 

df=pd.DataFrame()
df["Identity threshold"]=data1
df["Homology threshold"]=data2
df.index=[0,1,2,3,6,7,9,13,16,21,24,31]
df.plot(kind='bar',color={"Identity threshold": "Orange", "Homology threshold": "#20639b"},width=0.75,rot=0,figsize=(9,6))

plt.ylabel("Number of peptides that\n passed the threshold")
plt.xlabel("Labeling duration (Days)")

plt.savefig(f"mascotplot_muscle.jpeg",dpi=900, bbox_inches = "tight")
```

```python
from matplotlib import pyplot as plt
plt.figure(figsize=(40,15))
plt.rc('font', size=14) 

df=pd.DataFrame() 
df["Homology threshold"]=data2
df.index=[0,1,2,3,6,7,9,13,16,21,24,31]
df.plot(kind='bar',color="#20639b",width=0.55,rot=0,figsize=(9,6),legend=False) #, "Homology threshold": "#20639b"
plt.legend()

plt.ylabel("Number of PSMs ")
plt.xlabel("Labeling duration (Days)")
plt.gca().get_legend().remove()
plt.savefig(f"mascotplot_Homology_muscle.jpeg",dpi=900, bbox_inches = "tight")
```
