```python
import numpy as np
import pandas as pd
import os
import seaborn as sns
from matplotlib import pyplot as plt

import warnings

warnings.simplefilter("ignore")
```

```python
organ="liver"
# organ="heart"
# organ="kidney"
# organ="muscle"
time=16
```

```python
data_no_alignment=pd.read_csv(f"ElutionTime_compare_elutiontime_t0_t{time}_no_alignment_{organ}.csv")
print(data_no_alignment.shape)
data_no_alignment=data_no_alignment[(data_no_alignment.elutiontime_0<500) & (data_no_alignment.elutiontime_31<500)]
print(data_no_alignment.shape)

plt.rc('font', size=14) 
# plt.figure(figsize=(12,12))
plt.scatter(data_no_alignment.elutiontime_0,data_no_alignment.elutiontime_31,c='k',alpha=0.5)
plt.plot(data_no_alignment.elutiontime_0,data_no_alignment.elutiontime_0,c='r')

plt.xlabel("Elution time (min), Unlabeled sample (day 0)")
plt.ylabel("Elution time (min), Labeled sample (day 31)")

# plt.savefig(f"Elution_time_liver_t0_t31.jpeg",dpi=900, bbox_inches = "tight")
```
