```python
import numpy as np
import pandas as pd
import os
import seaborn as sns
from matplotlib import pyplot as plt
import warnings
import seaborn as sn
warnings.simplefilter("ignore")
```

```python
# organ="liver"
# organ="heart"
organ="kidney"
# organ="muscle"
time=16
```

```python
times=[1,6,16,31]
for t in times:
    plt.rc('font', size=14)
    plt.rcParams['figure.figsize'] = 10,7
    data_no_alignment=pd.read_csv(f"ElutionTime_compare_elutiontime_t0_t{t}_{organ}.csv")
    data_no_alignment=data_no_alignment[(data_no_alignment.elutiontime_0<500) & (data_no_alignment.elutiontime_31<500)]
    data_no_alignment["dif"]=abs(data_no_alignment.elutiontime_0-data_no_alignment.elutiontime_31)
    sn.kdeplot((data_no_alignment.elutiontime_0-data_no_alignment.elutiontime_31),shade=False,bw_adjust=7,linewidth=2,label=f'Day {t}')
    plt.xlim([-3,5])
    
plt.legend()

plt.savefig(f"Elution_time_{organ}_t0_t1_t6_t16_t31.jpeg",dpi=900, bbox_inches = "tight")
```
