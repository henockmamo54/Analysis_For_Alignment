```python

Finaldata["L1"]=[int(100*60*x) for x in Finaldata["L1"]]
Finaldata["L31"]=[int(100*60*x) for x in Finaldata["L31"]]
```

```python
chrom1=pd.read_csv("chrom1.txt")
chrom2=pd.read_csv("chrom31.txt")
chrom1["Time"]=[int(100*60*x) for x in chrom1.Time]
chrom2["Time"]=[int(100*60*x) for x in chrom2.Time]
print(chrom1.shape,chrom2.shape)
chrom1.head()
```

```python
merged=Finaldata.merge(chrom1,left_on="L1",right_on="Time")#[["L1","L31","Intensity"]]
merged=merged.merge(chrom2,left_on="L31",right_on="Time")
print(merged.shape)
merged.head(10)
```

```python
plt.figure(figsize=(10,6))
plt.rc('font', size=14) 
plt.plot(merged.L1/(100*60),merged.Intensity_x,c='k',label="Chromatogram 1")
plt.plot(merged.L31/(100*60),merged.Intensity_y,c='r',label="Chromatogram 31")

plt.ylabel("Abundance")
plt.xlabel("Time (min)")
plt.legend()
plt.savefig(f"withoutalignment.jpeg",dpi=900, bbox_inches = "tight")
```
