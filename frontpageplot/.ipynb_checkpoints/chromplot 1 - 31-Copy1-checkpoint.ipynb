```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
data=pd.read_csv("Align_Liver_HW_2h_L0_Liver_HW_2h_L1.txt",header=None)
L0=np.array(data.iloc[list(range(0,data.shape[0],2)),0])
L1=np.array(data.iloc[list(range(1,data.shape[0],2)),0])

tempdata=pd.DataFrame()
tempdata["L0"]=L0
tempdata["L1"]=L1
tempdata.to_csv("_Align_Liver_HW_2h_L0_Liver_HW_2h_L1.csv")

data=pd.read_csv("Align_Liver_HW_2h_L1_Liver_HW_2h_L2.txt",header=None)
L11=np.array(data.iloc[list(range(0,data.shape[0],2)),0])
L2=np.array(data.iloc[list(range(1,data.shape[0],2)),0])

tempdata1=pd.DataFrame()
tempdata1["L1"]=L11
tempdata1["L2"]=L2

tempdata1.to_csv("_Align_Liver_HW_2h_L1_Liver_HW_2h_L2.csv")
```

```python
Finaldata=pd.merge(tempdata,tempdata1,on="L1")
print(Finaldata.shape)
Finaldata.head()
```

```python
Finaldata.head()
```

```python

Finaldata["L1"]=[int(100*60*x) for x in Finaldata["L1"]]
Finaldata["L31"]=[int(100*60*x) for x in Finaldata["L31"]]
```

```python
chrom1=pd.read_csv("chrom1.txt")
chrom2=pd.read_csv("chrom31.txt")
chrom1["Time"]=[int(100*60*x) for x in chrom1.Time]
chrom2["Time"]=[int(100*60*x) for x in chrom2.Time]
print(chrom1.shape,chrom2.shape)
chrom1.head()
```

```python
merged=Finaldata.merge(chrom1,left_on="L1",right_on="Time")#[["L1","L31","Intensity"]]
merged=merged.merge(chrom2,left_on="L31",right_on="Time")
print(merged.shape)
merged.head(10)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
merged.Intensity_x=scaler.fit_transform(np.reshape(np.array(merged.Intensity_x), (-1, 2))).ravel()
merged.Intensity_y=scaler.fit_transform(np.reshape(np.array(merged.Intensity_y), (-1, 2))).ravel()
merged.head()
```

```python
plt.figure(figsize=(10,6))
plt.rc('font', size=14) 
plt.plot(merged.L1/(100*60),merged.Intensity_x,c='k',label="Chromatogram 1")
plt.plot(merged.L31/(100*60),merged.Intensity_y,c='r',label="Chromatogram 31")

plt.ylabel("Abundance")
plt.xlabel("Time (min)")
plt.legend()
plt.savefig(f"withoutalignment.jpeg",dpi=900, bbox_inches = "tight")
```
