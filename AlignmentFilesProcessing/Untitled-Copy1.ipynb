```python
ret1=[]
for i in tempscandata.exp1:
    if(int(i)!=-1):
        try:
            ret1.append([int(i),float(exp1[exp1.sn==int(i)].rt)])
        except:
            continue 
    else:
        continue 
        
ret1=pd.DataFrame(ret1,columns=["scan","ret"])

ret2=[]
for i in tempscandata.exp2:
    if(int(i)!=-1):
        try:
            ret2.append([int(i),float(exp2[exp2.sn==int(i)].rt)])
        except:
            continue 
    else:
        continue 
        
ret2=pd.DataFrame(ret2,columns=["scan","ret"])


ret3=[]
for i in tempscandata.exp3:
    if(int(i)!=-1):
        try:
            ret3.append([int(i),float(exp3[exp3.sn==int(i)].rt)])
        except:
            continue 
    else:
        continue 
        
ret3=pd.DataFrame(ret3,columns=["scan","ret"])


ret4=[]
for i in tempscandata.exp4:
    if(int(i)!=-1):
        try:
            ret4.append([int(i),float(exp4[exp4.sn==int(i)].rt)])
        except:
            continue 
    else:
        continue 
        
ret4=pd.DataFrame(ret4,columns=["scan","ret"])



ret12=[]
for i in tempscandata.exp12:
    if(int(i)!=-1):
        try:
            ret12.append([int(i),float(exp12[exp12.sn==int(i)].rt)])
        except:
            continue 
    else:
        continue 
        
ret12=pd.DataFrame(ret12,columns=["scan","ret"])
```

```python
temp=ret1.merge(ret12,on="scan")
temp1_2=ret1.merge(ret2,on="scan")
temp1_3=ret1.merge(ret3,on="scan")
temp1_4=ret1.merge(ret4,on="scan")
temp1_12=ret1.merge(ret12,on="scan")
temp2_12=ret2.merge(ret12,on="scan")
temp.head()
```

```python
from matplotlib import pyplot as plt
plt.figure(figsize=(16,8))
plt.rc('font', size=20) 
# plt.scatter(temp.ret_x,temp.ret_y)
 
plt.scatter(temp1_2.ret_x,abs(-temp1_2.ret_x+temp1_2.ret_y),alpha=0.5,label=" L1,L2")
plt.scatter(temp1_3.ret_x,abs(-temp1_3.ret_x+temp1_3.ret_y),alpha=0.5,label=" L1,L3")
plt.scatter(temp1_4.ret_x,abs(-temp1_4.ret_x+temp1_4.ret_y),alpha=0.5,label=" L1,L4")
plt.scatter(temp1_12.ret_x,abs(-temp1_12.ret_x+temp1_12.ret_y),alpha=0.5,label=" L1,L12")
plt.scatter(temp2_12.ret_x,abs(-temp2_12.ret_x+temp2_12.ret_y),alpha=0.5,label=" L2,L12")
plt.legend()
plt.ylabel("Abs. Retention time deviation (seconds)")
plt.xlabel("Retention time (seconds)")


plt.savefig('tempplot.jpeg',dpi=900, bbox_inches = "tight")
```
