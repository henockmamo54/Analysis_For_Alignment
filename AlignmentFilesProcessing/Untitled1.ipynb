```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
plt.figure(figsize=(16,6))
plt.rc('font', size=20) 

data=pd.read_csv("Align_Liver_HW_2h_L0_Liver_HW_2h_L1.txt",header=None)
L0=np.array(data.iloc[list(range(0,data.shape[0],2)),0])
L1=np.array(data.iloc[list(range(1,data.shape[0],2)),0])
plt.hlines(0,xmin=min(data.iloc[:,0]),xmax=max(data.iloc[:,0]),color='b')
plt.scatter(L0,(L0-L1),c='k',alpha=0.5)

plt.ylabel("Retention time deviation (min) [T0,T1]")
plt.xlabel("Retention time (min)")
print(L0.shape,L1.shape)
```
